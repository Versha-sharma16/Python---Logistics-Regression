# Question 2: What are the key assumptions of Simple Linear Regression?

Ans - Key Assumptions of Simple Linear Regression (SLR)

Simple Linear Regression is based on several important assumptions that ensure the validity and reliability of the model‚Äôs estimates and predictions. These assumptions relate to the relationship between the independent variable (X) and the dependent variable (Y), as well as the behavior of the residuals (errors).

 1. Linearity

- The relationship between the independent variable (X) and the dependent variable (Y) must be linear.
- This means that a change in X results in a proportional change in Y.
- Graphically: The data points should form a straight-line pattern when plotted.

Example:
If advertising expenditure doubles, sales should approximately double ‚Äî not increase exponentially or stay constant.

 2. Independence of Errors

- The residuals (errors) should be independent of each other.
- There should be no correlation between consecutive error terms.
- This is especially important in time-series data.

Example:
Errors in one month‚Äôs sales prediction should not be related to the next month‚Äôs errors.

 3. Homoscedasticity (Constant Variance of Errors)

- The residuals should have constant variance across all levels of X.
- In other words, the spread of residuals should remain roughly the same for all predicted values.

If violated:
You have heteroscedasticity, which can make significance tests unreliable.

Visual check:
A plot of residuals vs. predicted values should show a random scatter, not a funnel or cone shape.

 4. Normality of Errors

- The residuals (errors) should be approximately normally distributed.
- This assumption is important for making valid inferences (e.g., confidence intervals, hypothesis tests).

Check:
Plot a histogram or Q-Q plot of residuals ‚Äî it should look roughly bell-shaped.

 5. No (or Minimal) Measurement Error in X

- The independent variable (X) is assumed to be measured without error.
- If X has large measurement errors, the regression estimates may be biased and inefficient.

 6. No Significant Outliers

- Outliers can have a strong influence on the regression line, distorting slope and intercept estimates.
- It‚Äôs important to identify and handle them appropriately.

#Question 5: What is the method of least squares in linear regression?
Ans - The method of least squares in linear regression is a statistical technique used to find the best-fit line for a set of data points by minimizing the sum of the squared differences (residuals) between the observed and predicted values. This is done by minimizing the sum of the squares of the vertical distances of the data points from the line. The goal is to find a line that represents the data with the least amount of error, resulting in an equation that minimizes these squared errors.

###How it works
1. Plot the data: The independent variable is plotted on the x-axis and the dependent variable on the y-axis, creating a scatter plot.
2. Find the line of best fit: The method finds the line that best represents the overall trend of the points.
3. Calculate residuals: For each data point, the residual is the vertical distance between the actual y-value and the y-value predicted by the line.
4. Square the residuals: Each residual is squared. This ensures that all errors are positive and that larger errors are penalized more heavily.
5. Sum the squared residuals: The squared residuals are added together to get a single value representing the total squared error for the line.
6. Minimize the sum: The method then finds the specific line (defined by its slope and y-intercept) that makes this sum of squared residuals as small as possible.

What is the resulting equation?

- The final result is a linear equation, often written as \(y=mx+b\), where \(m\) is the slope and \(b\) is the y-intercept of the best-fit line.
- This equation can then be used to make predictions about the dependent variable (\(y\)) for new values of the independent variable (x).

Question 6: What is Logistic Regression? How does it differ from Linear Regression?

Ans - Logistic regression is a statistical method for predicting the probability of a categorical outcome, while linear regression is used for predicting a continuous numerical outcome. The main differences lie in their use cases (classification vs. regression), output type (categorical vs. continuous), and the underlying mathematical model (sigmoid curve vs. straight line).

###Logistic Regression

- What it is: A classification algorithm used to predict the probability of a binary outcome (e.g., yes/no, true/false, spam/not spam).
-- How it works: It uses a logistic (sigmoid) function to map the relationship between input variables and a continuous output to a probability between 0 and 1.
- Output: A probability value that is then used to classify the outcome based on a threshold (e.g., above 0.5 is "yes," below 0.5 is "no").
- Example: Predicting whether a customer will click an ad based on their demographics.
###Linear Regression
- What it is: A regression algorithm used to predict a continuous output variable.
- How it works: It models the relationship between independent variables and a dependent variable as a straight line.
- Output: A continuous numerical value, such as a price, temperature, or score.
- Example: Predicting the price of a house based on its size, location, and features.

#Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept. (Include your Python code and output in the code box below.)

Ans -

```python
# Import required libraries
from sklearn.linear_model import LinearRegression
import numpy as np

# Example dataset
# X = independent variable (e.g., house size in sq ft)
# y = dependent variable (e.g., house price in $)
X = np.array([500, 700, 1000, 1200, 1500]).reshape(-1, 1)
y = np.array([150000, 200000, 250000, 300000, 350000])

# Create the Linear Regression model
model = LinearRegression()

# Fit the model
model.fit(X, y)

# Print the slope (coefficient) and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

#Question 10: How do you interpret the coefficients in a simple linear regression model?

Ans - In a simple linear regression model, which has the form:


y=a+bX

where:

y = dependent variable

X = independent variable

a = intercept

b = slope (coefficient)

the coefficients are interpreted as follows:

1. Intercept (a)

- This is the predicted value of y when X=0.

- It represents the baseline or starting point of the regression line.

Example: If predicting house prices based on size, an intercept of 54,936 means that theoretically, a house with 0 sq ft would have a price of $54,936 (though this is just a mathematical baseline, not realistic).

2. Slope / Coefficient (b)

- This represents the change in y for a one-unit increase in X.

- It shows the direction and strength of the relationship between X and Y:

Positive b ‚Üí as X increases, Y increases

Negative b ‚Üí as X increases, Y decreases

- Example: A slope of 199.04 means that for each additional 1 sq ft of house size, the predicted price increases by $199.
